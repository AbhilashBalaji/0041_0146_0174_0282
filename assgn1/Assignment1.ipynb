```python
get_ipython().magic('matplotlib inline')
import matplotlib.pyplot as plt
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import seaborn as sns
```

```python
df = pd.read_csv('lok_sabha_2019_winners_affidavit.csv')
```

```python
freq, bins, patches = plt.hist(df.Age)  #fetching freq in each bin and bin values
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Histogram')
plt.show()
```

```python
plt.scatter(df.Liabilities_num, df.Assets_num)
plt.xlabel('Assets')
plt.ylabel('Liabilities')
plt.title('Assets vs Liabilities')
plt.show()
```

```python
df.Education.value_counts().plot(kind="bar")
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Bar Plot of Education Level of Candidates')
plt.show()
```

```python
plt.scatter(df.Criminal_Case, df.Age)
plt.xlabel('Criminal_Case')
plt.ylabel('Age')
plt.title('Criminal_Case vs Age')
plt.show()
```
